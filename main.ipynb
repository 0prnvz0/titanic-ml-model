```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error

train_data=pd.read_csv(r'D:\ml\practice\titanic\train.csv')
test_data=pd.read_csv(r'D:\ml\practice\titanic\test.csv')
```

```python
train_data.describe()
```

```python
train_data.dropna(axis=0)
```

```python
train_Y=train_data['Survived']
features=['Pclass','SibSp','Parch','Age','Fare']    
train_X=train_data[features]
model=DecisionTreeRegressor(random_state=0)
model.fit(train_X,train_Y)
```

```python
predictions=model.predict(train_X)
```

```python
mae = mean_absolute_error(train_Y, predictions)
print(f"Mean Absolute Error: {mae:.4f}")
```
